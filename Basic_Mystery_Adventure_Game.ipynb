```python
import time

def print_slow(text):
    """Prints text slowly for a dramatic effect."""
    for char in text:
        print(char, end='', flush=True)
        time.sleep(0.03)
    print()

def game_intro():
    print_slow("You wake up in a cold, dark room.")
    print_slow("The air is thick with the scent of decay.")
    print_slow("You have no memory of how you got here.")
    print_slow("Your only goal is to escape the mansion...")

def room1():
    print_slow("\nYou are in the main hall.")
    print_slow("There are three doors in front of you:")
    print_slow("1. A red door to the left.")
    print_slow("2. A blue door straight ahead.")
    print_slow("3. A green door to the right.")

    choice = input("Which door do you choose? (1/2/3): ")

    if choice == "1":
        print_slow("\nYou open the red door and step into a dimly lit library.")
        library()
    elif choice == "2":
        print_slow("\nYou open the blue door and enter a dusty dining room.")
        dining_room()
    elif choice == "3":
        print_slow("\nYou open the green door and find yourself in a creepy bedroom.")
        bedroom()
    else:
        print_slow("\nInvalid choice! Try again.")
        room1()

def library():
    print_slow("\nThe library is filled with ancient books.")
    print_slow("You notice a book glowing faintly on a pedestal.")
    print_slow("Do you:")
    print_slow("1. Open the book.")
    print_slow("2. Ignore it and search for an exit.")

    choice = input("What do you do? (1/2): ")

    if choice == "1":
        print_slow("\nYou open the book and are transported to another dimension!")
        print_slow("Game Over. You are lost forever.")
    elif choice == "2":
        print_slow("\nYou find a hidden passage behind a bookshelf.")
        print_slow("You escape the mansion!")
        print_slow("Congratulations! You win!")
    else:
        print_slow("\nInvalid choice! Try again.")
        library()

def dining_room():
    print_slow("\nThe dining room is filled with cobwebs and broken furniture.")
    print_slow("You see a key on the table and a window that might lead outside.")
    print_slow("Do you:")
    print_slow("1. Take the key.")
    print_slow("2. Try to climb out the window.")

    choice = input("What do you do? (1/2): ")

    if choice == "1":
        print_slow("\nYou take the key and hear a loud creak.")
        print_slow("A trapdoor opens beneath you, and you fall into darkness.")
        print_slow("Game Over. You are never seen again.")
    elif choice == "2":
        print_slow("\nYou climb out the window and escape the mansion!")
        print_slow("Congratulations! You win!")
    else:
        print_slow("\nInvalid choice! Try again.")
        dining_room()

def bedroom():
    print_slow("\nThe bedroom is eerily silent.")
    print_slow("You see a mirror and a closet.")
    print_slow("Do you:")
    print_slow("1. Look into the mirror.")
    print_slow("2. Open the closet.")

    choice = input("What do you do? (1/2): ")

    if choice == "1":
        print_slow("\nYou look into the mirror and see a ghostly figure behind you.")
        print_slow("You are paralyzed with fear and cannot move.")
        print_slow("Game Over. The ghost claims you.")
    elif choice == "2":
        print_slow("\nYou open the closet and find a secret passage.")
        print_slow("You escape the mansion!")
        print_slow("Congratulations! You win!")
    else:
        print_slow("\nInvalid choice! Try again.")
        bedroom()

def main():
    game_intro()
    room1()

# Run the game
main()
```
